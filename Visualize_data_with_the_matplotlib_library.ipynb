## Multiple ranges
You can show several ranges on the same plot, each with different colors and styles.

Lines are blue by default. You specify colors with the <a href="http://www.mathworks.com/help/matlab/ref/colorspec.html" target="_blank" rel="noopener noreferrer">MathWorks ColorSpec</a>. 

You can use either the short name or long name of the color:

```
RGB Value Short Name Long Name
[1 1 0]   y          yellow
[1 0 1]   m          magenta
[0 1 1]   c          cyan
[1 0 0]   r          red
[0 1 0]   g          green
[0 0 1]   b          blue
[1 1 1]   w          white
[0 0 0]   k          black
```

For example, `color='r'` sets the color to red.

Lines are solid by default. You specify shapes, such as dashes, squares, and triangles, with the <a href="http://www.mathworks.com/help/matlab/ref/linespec.html" target="_blank" rel="noopener noreferrer">LineSpec property</a>. To combine a custom color and a custom line shape, you must use the short name of the color.

Create a plot with multiple ranges that have custom colors and shapes.

````python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots()

import numpy as np

# evenly sample time at 200ms intervals
t = np.arange(0., 5., 0.2)

# magenta dashes (--) for t, 
# green squares (s) for t^2
# cyan triangles (^) for t^3
plt.plot(t, t, 'm--', t, t**2, 'gs', t, t**3, 'c^')
plt.show()
````

### Titles and labels

You can add a title to each axis in a figure. To set the title, use the `set_title` method in the axes instance:

```
ax.set_title("title")
```

Similarly, use the methods `set_xlabel` and `set_ylabel` to set the labels of the X and Y axes:

```
ax.set_xlabel("x")
ax.set_ylabel("y")
```

Create two plots and set the figure titles, axis labels, and legends.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = np.linspace(0, 5, 10)
y = x ** 2
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
    
fig.tight_layout()
````

<a id="axes2"></a>
### Legends

Legends for curves in a figure can be added in two ways. One method is to use the MATLAB API `legend` function of the axis object to pass a list of legend texts for the previously defined curves:

```
ax.legend(["curve1", "curve2", "curve3"])
```

Be careful when you use this method: If you add or remove curves from the figure, you must update the `legend` function accordingly.

A more flexible method is to include the `label="label text"` keyword argument for each curve, and include the `legend` method without label name arguments: 

```
ax.plot(x, x**2, label="curve1")
ax.plot(x, x**3, label="curve2")
ax.legend()
```

When you add or remove curves to or from the figure, the legend is automatically updated.

The `legend` function takes an optional keyword argument, `loc`, that you can use to specify the location of the legend. The allowed values of `loc` are numerical codes for the various locations. See <a href="http://matplotlib.org/users/legend_guide.html#legend-location" target="_blank" rel="noopener noreferrer">Legend location</a> for details. 

Some of the most common `loc` values are:

```
ax.legend(loc=0) # let matplotlib decide the optimal location
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner
```
Create a plot with a legend.

````python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots()

import numpy as np

x = np.linspace(0, 5, 10)

ax.plot(x, 2*x**2, label="y = 2x^2")
ax.plot(x, x**3, label="y = x^3")
ax.legend(loc=2); # upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title');
````

<a id="axes3"></a>
### Limits 
You can set limits on the range of values for axes.

To configure the ranges of the axes, use the `set_ylim` and `set_xlim` functions of the axis object, or use `axis('tight')` function to automatically display "tightly fitted" axis ranges.

Create three plots: one with default ranges, one with tightly-fitted ranges, and one with custom ranges.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
x = np.linspace(0, 5, 10)
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Default Axes Range")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Tight Axes")

axes[2].plot(x, np.power(x,2), x, np.power(x,3))
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Custom Axes Range");
````

## Multiple plots

Use the `subplot()` function to support multiple plots in the same figure.  
The `subplot()` function parameter has the following arguments:

 - number of rows (`nrows`)
 - number of columns (`ncols`)
 - the plot number (`plot_num`), where the value can be 1 to `nrows * ncols`. 

You don't need to include a delimiter between the arguments if `(nrows*ncols)<10`, but you can include commas for readability. For example, `subplot(211)` is equivalent to `subplot(2,1,1)`. If `(nrows*ncols)=>10`, you must include commas between arguments. 

In a figure that has 2 rows and 1 column, `subplot(211)` refers to the **first** plot, and `subplot(212)` refers to the **second** plot.

You can create any number of subplots and axes. If you want to place axes manually, instead of on a rectangular grid, use the `axes()` function to specify the location as `axes([left, bottom, width, height])`, where all values are in fractional (0 to 1) coordinates. See an example at <a href="http://matplotlib.org/examples/pylab_examples/axes_demo.html" target="_blank" rel="noopener noreferrer">axes_demo.py</a>.

Create a figure with two plots.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# define a function to plot, needs numpy
def f(t): 
    return np.exp(-t) * np.cos(2.5*np.pi*t)

t1 = np.arange(0.0, 6.0, 0.2)
t2 = np.arange(0.0, 6.0, 0.1)

# '1' parameter is optional, specifying no parameter
# also creates a single plot
plt.figure(1) 
plt.subplot(211)
plt.plot(t1, f(t1), 'yo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2.5*np.pi*t2), 'r--')
plt.show()
````

<a id="fonts1"></a>
### Include special characters with LaTeX
You can include special characters because matplotlib includes support for LaTeX. To render text as LaTeX, surround the text with dollar sign characters ($), for example: `"$y=x^3$"`.

**Note about using backslash (\\):** In LaTeX, you use the backslash to produce the symbols, for example: `$\alpha$`. However, the backslash already has a meaning in Python strings: the escape code character. To properly render the LaTex code, use "raw" text strings. To make a raw text string, add the letter `r` in front of a text string, for example: `r"\alpha"` or `r'\alpha'` instead of `"\alpha"` or `'\alpha'`.

Create a plot with LaTex that uses the alpha character.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 5, 10)
ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18)
ax.set_title('Title');
````

<a id="fonts2"></a>
### Customize fonts with the rcParams variable
You can change the global font properties, which applies to all text elements in a figure (tick labels, axis labels and titles, legends, and so on), by updating the `rcParams` variable.

Include this code to change the font size to 12 and the font family to serif:

```
import matplotlib as matplotlib
matplotlib.rcParams.update({
    'font.size': 12, 
    'font.family': 'serif'
})
```
Create a plot custom font properties.

````python
import matplotlib as matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 5, 10)
matplotlib.rcParams.update({'font.size': 12, 'font.family': 'serif'})

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('Title');
````

<a id="fonts3"></a>
### Use STIX fonts with the rcParams variable
A good choice of global fonts are the <a href="http://www.stixfonts.org/" target="_blank" rel="noopener noreferrer">STIX fonts</a>. You specify STIX fonts with the `rcParams` variable.

Create a plot that uses STIX fonts.

````python
import matplotlib as matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 5, 10)
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('My Cool Title');
````

<a id="stacked"></a>
## Stacked bar chart
To create a stacked bar graph, include the `bottom` argument in the `bar()` function. The bottom data represents the bottom coordinate of each data range. 

Create a stacked bar chart with the data for men plotted below the data for women.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

N = 5
menMeans   = (40, 36, 31, 36, 28)
womenMeans = (35, 22, 24, 30, 25)
menStd     = (2, 3, 4, 1, 2)
womenStd   = (3, 5, 2, 3, 3)
ind = np.arange(N)  # the x locations for the groups
width = 0.35        # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans,   width, color='b')
p2 = plt.bar(ind, womenMeans, width, color='m',
             bottom=menMeans) # specify what's going to the bottom

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind+width/2., ('G1', 'G2', 'G3', 'G4', 'G5') )
plt.yticks(np.arange(0,91,10))
plt.legend( (p1[0], p2[0]), ('Men', 'Women') )

plt.show()
````

## Error bars
Use the `errorbar()` function and its `xerr` and `yerr` arguments to specify error margins. The error bar is a vertical line that looks like this `|----|` and indicates a margin of error above and below the y values.

Create a bar chart with error bars.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = [1, 2, 3, 4]
y = [20, 21 , 20.5, 20.9]

#define how much the error
y_error = [.12, .13, .2, .1]

#plot data
plt.bar(x,y, color="yellowgreen", align='center')
plt.errorbar(x,y, yerr=y_error, linestyle="None", 
             marker="None", color="green")

#configure  X axis
plt.xlim(0.5, 4.5)
plt.xticks([1,2,3,4])
#configure Y axis
plt.ylim(19.8,21.2)
plt.yticks([20,21,20.5,20.8])

#title and labels
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Error Margin Bar")

#show plot
plt.show()
````

Create a line plot with error bars.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = [1, 2, 3, 4]
y = [20, 21 , 20.5, 20.9]

#errors
y_error = [.12, .13, .2, .1]

#plot data
plt.plot(x,y, linestyle="dashed", marker="o", color="green")
plt.errorbar(x,y, yerr=y_error, linestyle="None", 
             marker="None", color="green")

#configure  X axis
plt.xlim(0.5, 4.5)
plt.xticks([1,2,3,4])
#configure Y axis
plt.ylim(19.8,21.2)
plt.yticks([20,21,20.5,20.8])

#title and labels
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Error Margin Line")
#show plot
plt.show()
````

## Histogram
You can use a histogram to display probability data. Use the `hist()` function to create a histogram. The `bins` parameter specifies the number of bars to be displayed.  Set the `cumulative=True` parameter to create a cumulative histogram, where all values sum to 100%.

Create a subplot with a default histogram and a subplot with a cumulative histogram.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n, bins=25, color = 'm')
axes[0].set_title("Default histogram")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=40, color='c')
axes[1].set_title("Cumulative detailed histogram")
axes[1].set_xlim((min(n), max(n)));
````

## Pie chart
You can create a pie chart with the `pie()` function. By default, this function chooses its own colors, and does not display legends or label names. You can set colors, legends, labels, the starting angle, shadowing, and whether any section is exploded, with the `pie()` function parameters.

The default starting angle is 0, which starts the first slice on the x-axis. If you set `startangle=90`, everything is rotated counter-clockwise by 90 degrees, so that the plotting starts on the positive y-axis.

Create a pie chart with custom colors, labels, and starting angle.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

data = [5,20,15,40,10,10]  
labels = ['A','B','C','D','E','F'] 
plt.figure(1, figsize=(6,6))  # make it square
colors = ['yellowgreen', 'gold', 'lightskyblue', 
          'lightcoral', 'green', 'red']
plt.pie(data, labels=labels, colors= colors, startangle=90)

plt.show()
````

Create a pie chart with labels, an exploded view, a shadow effect, a custom starting angle, and a title.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.figure(1, figsize=(6,6))

# The slices are ordered and plotted counter-clockwise.
# The second slice is exploded and shadowed.
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
fracs = [15, 30, 45, 10]
explode=(0, 0.05, 0, 0)

plt.pie(fracs, explode=explode, labels=labels,
                autopct='%1.1f%%', shadow=True, startangle=90)

plt.title('Raining Hogs and Dogs', bbox={'facecolor':'0.8', 'pad':5})

plt.show()
````

## Polar plot
You can create these kind of polar plots by including the `polar=True` argument:

 - [Polar line plot](#polar1)
 - [Polar bar plot](#polar2)
 - [Polar scatter plot](#polar3)

<a id="polar1"></a>
### Polar line plot
Create a polar line plot.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)
t = np.linspace(0, 1.5 * np.pi, 100)
ax.plot(t, t, color='red', lw=3);
````

<a id="polar2"></a>
### Polar bar plot
Create a polar bar plot.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

N = 21
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)

ax = plt.subplot(111, polar=True) #polar=True for polar plot
bars = ax.bar(theta, radii, width=width, bottom=0.0)

# Use custom colors and opacity
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.jet(r / 10.))
    bar.set_alpha(0.5)

plt.show()
````

<a id="polar3"></a>
### Polar scatter plot
Create a polar scatter plot.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

N = 50
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2 * np.random.rand(N)
colors = theta

ax = plt.subplot(111, polar=True)
c = plt.scatter(theta, r, c=colors, s=area, cmap=plt.cm.hsv)
c.set_alpha(0.75)

plt.show()
````

## 3D Plots
To use 3D graphics in matplotlib, create an instance of the `Axes3D` class. You can add 3D axes to a matplotlib figure canvas in the same way as you add 2D axes; or, more conveniently, by passing a `projection='3d'` argument to the `add_axes` or `add_subplot` functions.

You must import the `Axes3D` class:

```
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

You can create these types of 3D plots:

 - [3D surface plot](#3dplots1)
 - [3D wireframe plot](#3dplots2)
 - [3D contour plot](#3dplots3)

<a id="3dplots1"></a>
### 3D surface plot
You create a surface with the `plot_surface()` function. You can customize the stride (the space between each contour line extension), the linewidth of that line, and many more properties with the appropriate function parameters.

By default, a surface will be colored in shades of a solid color, but it also supports color mapping by supplying the `cmap` argument. Some arguments are `coolwarm` and `hot`.

Here's a list of the most common arguments for 3D surface plots:

```
X, Y, Z      Data values as 2D arrays
rstride      Array row stride (step size), defaults to 10
cstride      Array column stride (step size), defaults to 10
color        Color of the surface patches
cmap         A colormap for the surface patches.
facecolors   Face colors for the individual patches
norm         An instance of Normalize to map values to colors
vmin         Minimum value to map
vmax         Maximum value to map
shade        Whether to shade the facecolors
```

The entire list of parameters is here: <a href="http://matplotlib.org/mpl_toolkits/mplot3d/api.html" target="_blank" rel="noopener noreferrer">mplot3d API</a>.

You can also use regular matplotlib parameters for axes, titles, colors, and so on.

Create two 3D surface subplots, one with color grading and a color bar.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14,6))

alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p)*np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T

# `ax` is a 3D-aware axis instance because of the 
# projection='3d' keyword argument to add_subplot
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=3, cstride=4, linewidth=0)

# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                    cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
````

<a id="3dplots2"></a>
### 3D wireframe
Use the `plot_wireframe()` function to create a wireframe surface. 

These are the most common arguments for 3D wireframe plots:
```
X, Y, Z      Data values as 2D arrays
rstride      Array row stride (step size), defaults to 1
cstride      Array column stride (step size), defaults to 1
```

You can also use regular matplotlib parameters for axes, titles, colors, and so on.

Create a 3D wireframe plot.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D

alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p)*np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1, 1, 1, projection='3d')

p = ax.plot_wireframe(X, Y, Z, rstride=3, cstride=3, color='c')
````

<a id="3dplots3"></a>
### 3D contour plot
Add the `contourf()` function to an axis to have a projection of the 3D graph on the specified axis.

Create a 3D surface plot with color mapping and a contour on the Z axis.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt( X**2 + Y **2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap = plt.cm.hot)
ax.contourf(X, Y, Z , zdir='z', offset= -2, cmap=plt.cm.hot)
ax.set_zlim(-2,2)
plt.show()
````
